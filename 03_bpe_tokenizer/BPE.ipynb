```python
texto = """Era uma vez uma doce menina que vivia com sua mãe próximo a uma grande floresta. Ela sempre usava uma capa com capuz vermelho que sua avó lhe dera, e por isso todos a chamavam de Chapeuzinho Vermelho. 
Um dia, sua mãe pediu que ela levasse uma cesta com bolos, doces e frutas para a vovó, que estava doente e morava do outro lado da floresta. A mãe alertou: "Vá direto para a casa da vovó e não fale com estranhos". 
No caminho, distraída, Chapeuzinho encontrou o lobo mau. Sem saber do perigo, disse a ele que ia visitar sua avó. O lobo, astuto, sugeriu que ela colhesse flores, enquanto ele corria pelo caminho mais rápido até a casa da vovó. 
O lobo chegou, enganou a vovó e a engoliu. Depois, vestiu as roupas dela e deitou-se na cama esperando a menina. Quando Chapeuzinho chegou, achou a avó estranha:
— Vovó, que orelhas tão grandes! — disse ela.
— É para te ouvir melhor — respondeu o lobo.
— Vovó, que olhos tão grandes! — disse ela.
— É para te ver melhor.
— Vovó, que boca enorme é essa?
— É para te devorar! — gritou o lobo, que saltou e engoliu a Chapeuzinho. 
Um caçador que passava por ali ouviu o ronco do lobo e, desconfiado, entrou na casa. Ele abriu a barriga do lobo, salvando a vovó e a Chapeuzinho, que prometeram nunca mais desobedecer aos conselhos da mãe. """
```

```python
tokens = list(texto.encode('utf-8'))
```

```python
def get_stats(tokens):
    counts = {}
    for pair in zip(tokens, tokens[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts

get_stats(tokens)
```

```python
def merge(tokens, pair, new_id):
    new_tokens = []
    i = 0
    while i < len(tokens):
        if i < len(tokens) - 1 and (tokens[i], tokens[i + 1]) == pair:
            new_tokens.append(new_id)
            i += 2
        else:
            new_tokens.append(tokens[i])
            i += 1
    return new_tokens

teste = [1, 2, 3, 4, 1, 2, 1, 2, 5]
merge(teste, (1, 2), 100)
```

```python
## loop de treinamento
tokens = list(texto.encode('utf-8'))
vocab_size = 400
merges = {}

for i in range(vocab_size - 256):
    stats = get_stats(tokens)
    top_pair = max(stats, key=stats.get)
    new_id = 256 + i
    tokens = merge(tokens, top_pair, new_id)
    merges[top_pair] = new_id
    print(f"Merge {i+1}: {top_pair} -> {new_id} (Frequência: {stats[top_pair]})")
```

```python
def encode(text, merges):
    tokens = list(text.encode('utf-8'))
    
    while len(tokens) > 1:
        stats = get_stats(tokens)

        pair = min(stats, key=lambda p: merges.get(p, float('inf')))
        if pair not in merges:
            break
        tokens = merge(tokens, pair, merges[pair])

    return tokens

encoded = encode(texto, merges)
print("Encoded:", encoded)
print(len(texto))
print(len(encoded))
```

```python
def build_vocab(merges):
    vocab = {i: bytes([i]) for i in range(256)}
    for (a, b), new_id in merges.items():
        vocab[new_id] = vocab[a] + vocab[b]
    return vocab

vocab = build_vocab(merges)
```

```python
def decode(tokens, vocab):
    decoded_bytes = b''.join(vocab[token] for token in tokens)
    return decoded_bytes.decode('utf-8', errors='ignore')

decoded = decode(encoded, vocab)
print("Decoded:", decoded)
```

```python
import json

def save_tokenizer(file_path, merges):
    """Salva os merges em um arquivo JSON."""
    # Convertemos a tupla (id1, id2) em uma string "id1,id2" para o JSON
    serializable_merges = {f"{p1},{p2}": v for (p1, p2), v in merges.items()}
    
    with open(file_path, "w", encoding="utf-8") as f:
        json.dump(serializable_merges, f, indent=4)
    print(f"Tokenizador salvo em: {file_path}")

save_tokenizer("tokenizer.json", merges)
```

```python
from pprint import pprint

def load_tokenizer(file_path):
    """Carrega os merges e reconstrói o dicionário original."""
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    # Reconvertemos a string "id1,id2" de volta para a tupla (int, int)
    merges = {}
    for key, val in data.items():
        p1, p2 = map(int, key.split(","))
        merges[(p1, p2)] = val
        
    return merges

merges = load_tokenizer("tokenizer.json")
pprint(merges)
```

```python
def print_merges(merges, vocab):
    for (p1, p2), new_id in merges.items():
        print(f"({p1}, {p2}) -> {decode([new_id], vocab)}")

print_merges(merges, vocab)
```

```python
import re

# Padrão simplificado
SPLIT_PATTERN = r"""[a-zA-Zà-úÀ-Ú]+\s?|\d|\s|."""

chunks = re.findall(SPLIT_PATTERN, texto)
print(chunks)
```

```python
def get_stats_v2(chunks):
    counts = {}
    for chunk in chunks:
        # Conta pares apenas dentro de cada sub-sequência isolada
        for pair in zip(chunk, chunk[1:]):
            counts[pair] = counts.get(pair, 0) + 1
    return counts

def merge_v2(chunks, pair, replacement_id):
    new_chunks = []
    for chunk in chunks:
        new_chunk = []
        i = 0
        while i < len(chunk):
            if i < len(chunk) - 1 and (chunk[i], chunk[i+1]) == pair:
                new_chunk.append(replacement_id)
                i += 2
            else:
                new_chunk.append(chunk[i])
                i += 1
        new_chunks.append(new_chunk)
    return new_chunks
```

```python
## loop de treinamento
chunks = re.findall(SPLIT_PATTERN, texto)
tokens = [list(chunk.encode('utf-8')) for chunk in chunks]
vocab_size = 400
merges = {}

for i in range(vocab_size - 256):
    stats = get_stats_v2(tokens)
    top_pair = max(stats, key=stats.get)
    new_id = 256 + i
    tokens = merge_v2(tokens, top_pair, new_id)
    merges[top_pair] = new_id
    print(f"Merge {i+1}: {top_pair} -> {new_id} (Frequência: {stats[top_pair]})")
```

```python
vocab = build_vocab(merges)
print_merges(merges, vocab)
```

```python
encoded = encode(texto, merges)
print("Encoded:", encoded)
print(len(encoded))
```
