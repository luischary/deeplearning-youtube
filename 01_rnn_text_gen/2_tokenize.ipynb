```python
import pandas as pd
```

```python
df = pd.read_parquet("data/base_treino.pq")
df
```

```python
from tqdm import tqdm

vocab = set()
for r in tqdm(df.itertuples()):
    texto = r.texto
    for char in texto:
        vocab.add(char)
```

```python
print(vocab)
```

```python
print(len(vocab))
```

```python
tokenize_dict = {char: i for i, char in enumerate(vocab)}
detokenize_dict = {i: char for char, i in tokenize_dict.items()}

from pprint import pprint
pprint(tokenize_dict)
```

```python
import json
from pathlib import Path

Path("artifacts/tokenize_dict.json").write_text(json.dumps(tokenize_dict), encoding="utf8")
Path("artifacts/detokenize_dict.json").write_text(json.dumps(detokenize_dict), encoding="utf8")
```
