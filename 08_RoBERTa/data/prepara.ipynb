```python
import pandas as pd
```

```python
base = pd.read_csv('IMDB Dataset.csv')
base
```

```python
base["label"] = base.sentiment.map({"positive": 1, "negative": 0})
base.label.value_counts()
```

```python
split = ["valid" for _ in range(1000)] + ["test" for _ in range(5000)] + ["train" for _ in range(len(base) - 6000)]

base = base.sample(frac=1.0, replace=False, random_state=1)
base["split"] = split
base.split.value_counts()
```

```python
base = base.rename(columns={"review": "texto"})
```

```python
from tokenizer import BPETokenizer, SPECIAL_TOKENS

tokenizer = BPETokenizer(vocab_size=5_000, merges_path="C:/Users/luisc/local/projetos/videos/transformer_/artifacts/tokenizer_wiki_en_5k.json")
tokenized = tokenizer.batch_encode(base.texto.tolist())
```

```python
print(tokenizer.decode(tokenized[0]))
```

```python
base["tokens"] = tokenized
```

```python
base.to_parquet("base_imdb.pq")
```
